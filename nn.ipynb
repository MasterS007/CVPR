```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from tqdm import tqdm
```

```python
_DATA_SET = "C:\\Users\\munir\\Desktop\\cvpr\\CIFAR-10-images-master\\CIFAR-10-images-master\\train"
_CATEGORIES = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
_TRAIN_IMGS = []
for c in _CATEGORIES:
    path = os.path.join(_DATA_SET, c)
    #print(path)
    class_name = _CATEGORIES.index(c)
    for img in tqdm(os.listdir(path)): 
        try:
            img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
            _TRAIN_IMGS.append([img_arr, class_name])
        except Exception as e:
            pass

    
```
